```python
import os
import numpy as np
import pandas as pd

from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
model = SentenceTransformer(
    "sentence-transformers/all-mpnet-base-v2"
)
```

```python
hope_sentences = [

# Future
"The future can be better than today.",
"There is always another opportunity.",
"Tomorrow offers a fresh beginning.",
"Every sunrise brings new possibilities.",
"I believe better days are ahead.",
"Life continues to offer hope.",
"There is always another path.",
"New opportunities appear with time.",
"Every ending opens another beginning.",
"The future still holds promise.",

# Agency
"I can overcome difficult situations.",
"I believe in my abilities.",
"I choose to move forward.",
"I have the strength to continue.",
"I will not give up.",
"I can solve difficult problems.",
"My choices matter.",
"I control my future.",
"I believe in myself.",
"I can create positive change.",

# Healing
"Pain will eventually heal.",
"Time helps emotional wounds recover.",
"I become stronger after hardship.",
"Healing takes patience.",
"I forgive myself and move on.",
"Love helps people heal.",
"I recover little by little.",
"My heart can heal.",
"Recovery is possible.",
"I embrace healing.",

# Resilience
"I rise after every fall.",
"I learn from failure.",
"I continue despite setbacks.",
"I keep moving forward.",
"I never stop trying.",
"Every obstacle teaches something.",
"Challenges make me stronger.",
"I remain hopeful.",
"I stand back up.",
"I face adversity with courage.",

# Connection
"I am not alone.",
"Love gives people strength.",
"Someone cares about me.",
"Kindness changes lives.",
"We support one another.",
"Together we become stronger.",
"Friendship gives hope.",
"People help each other.",
"I feel connected.",
"Love makes healing possible."
]
```

```python
despair_sentences = [

# Hopelessness
"There is no hope left.",
"Nothing will ever improve.",
"The future feels empty.",
"I cannot see a way forward.",
"Everything feels pointless.",
"I have completely given up.",
"There is nothing left to believe in.",
"Tomorrow will be no different.",
"Nothing can change.",
"I see no future.",

# Helplessness
"I have no control.",
"I cannot change anything.",
"I feel powerless.",
"My efforts never matter.",
"I cannot escape this pain.",
"I am trapped forever.",
"I cannot find a solution.",
"I always fail.",
"I feel defeated.",
"I cannot continue.",

# Isolation
"I am completely alone.",
"No one understands me.",
"Nobody cares anymore.",
"I feel abandoned.",
"I have no one to rely on.",
"I am disconnected from everyone.",
"I feel invisible.",
"I have been forgotten.",
"I am isolated.",
"I have lost everyone.",

# Loss
"I have lost everything.",
"My dreams are gone.",
"Nothing remains.",
"Everything I loved has disappeared.",
"I cannot recover from this loss.",
"My future has vanished.",
"I have nothing left.",
"My life feels broken.",
"I cannot replace what I lost.",
"I feel empty.",

# Meaninglessness
"My life has no meaning.",
"Nothing matters anymore.",
"There is no purpose.",
"I exist without direction.",
"I see no reason to continue.",
"Everything feels meaningless.",
"There is no light ahead.",
"I have lost myself.",
"I cannot find happiness.",
"There is no reason to hope."
]
```

```python
hope_embeddings = model.encode(
    hope_sentences,
    convert_to_numpy=True,
    show_progress_bar=True
)

despair_embeddings = model.encode(
    despair_sentences,
    convert_to_numpy=True,
    show_progress_bar=True
)
```

```python
hope_vector = hope_embeddings.mean(axis=0)
hope_vector = hope_vector / np.linalg.norm(hope_vector)

despair_vector = despair_embeddings.mean(axis=0)
despair_vector = despair_vector / np.linalg.norm(despair_vector)
```
